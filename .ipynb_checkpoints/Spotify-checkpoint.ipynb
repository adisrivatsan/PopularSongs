## Trend of top songs in US in 2017

```python
us_songs_df = songs_df[songs_df.Region == 'us']
# us_songs_df = songs_df
streams_df = us_songs_df[['Date', 'Track Name', 'Streams', 'Position']]
streams_df = streams_df.groupby(['Date', 'Track Name']).first().reset_index()
stream_num = streams_df.pivot(index = 'Date', columns = 'Track Name', values = 'Streams')
pos_num = streams_df.pivot(index = 'Date', columns = 'Track Name', values = 'Position')
us_unique_song = us_songs_df[us_songs_df['Position'] == 1]['Track Name'].unique()

us_unique_song
```

```python
def draw_stream_pos(song):
    song_stream = stream_num[song]
    song_pos = pos_num[song]
    ind = song_stream.notnull()
    song_stream = song_stream.loc[ind]
    song_pos = song_pos.loc[ind]
    first = (song_pos.values == 1)
    not_first = ~first
    f, ax = plt.subplots(1,2,figsize = (12,4))
    ax[0].scatter(song_stream.index[not_first], song_stream.values[not_first], alpha = 0.4, color = 'grey')
    ax[0].scatter(song_stream.index[first], song_stream.values[first], color = 'red')
    ax[1].scatter(song_pos.index[not_first], song_pos.values[not_first], alpha = 0.4, color = 'grey')
    ax[1].scatter(song_pos.index[first], song_pos.values[first], color = 'red')
    ax[1].invert_yaxis()
    ax[1].set_title('Top1 Number : '+ str((song_pos.values == 1).sum()))
    ax[0].set_title(song)
    plt.show()
```

```python
for song in us_unique_song:
    draw_stream_pos(song)
```

## Trend of top songs Globally which were in 1st position in the US in 2017

```python
global_songs_df = songs_df[songs_df.Region == 'global']
streams_df = global_songs_df[['Date', 'Track Name', 'Streams', 'Position']]
streams_df = streams_df.groupby(['Date', 'Track Name']).first().reset_index()
stream_num = streams_df.pivot(index = 'Date', columns = 'Track Name', values = 'Streams')
pos_num = streams_df.pivot(index = 'Date', columns = 'Track Name', values = 'Position')
```
